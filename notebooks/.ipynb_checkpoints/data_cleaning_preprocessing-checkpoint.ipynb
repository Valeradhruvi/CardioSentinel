# OUTLIER DETECTION & HANDLING

This dataset contains major outliers in ap_hi and ap_lo.

Detect Outliers using IQR

```python
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    outliers = df[(df[column] < lower) | (df[column] > upper)]
    return outliers

outlier_columns = ['height', 'weight', 'ap_hi', 'ap_lo', 'age']
for col in outlier_columns:
    outliers = detect_outliers(df, col)
    print(f"Outliers in {col}: {len(outliers)}")
```

Fix Outliers (Capping / Winsorization)

We fix by setting extreme values to acceptable range:

```python
def cap_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    df[column] = df[column].clip(lower, upper)

for col in outlier_columns:
    cap_outliers(df, col)
df.describe()
```

# FEATURE ENGINEERING

Convert Age from Days to Years

```python
df['age_years'] = (df['age'] / 365).astype(int)
df.drop('age', axis=1, inplace=True)
df.head()
```

Create BMI

```python
df['BMI'] = df['weight'] / ( (df['height']/100) ** 2 )
df.head()
```

# FEATURE SCALING

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scale_cols = ['height','weight','ap_hi','ap_lo','BMI']
df[scale_cols] = scaler.fit_transform(df[scale_cols])
df.head()
```
